## Feature Engineering

```python
def make_date(df):
    new_date = ""
    year = df['iyear']
    month = df['imonth']
    day = df['iday']
    
    if month < 1:
        month = 1
        
    if day < 1:
        day = 1
    
    new_date = f"{year}-{month}-{day}"
    
    return new_date
```

```python
new_df = new_df.assign(
    date=new_df.apply(make_date, axis=1)
)
```

The date column of `new_df` can be set to index to aid time series analyses

```python
new_df.assign(date=new_df["date"].astype("datetime64[ns]")).set_index("date").head()
```

```python
new_df.drop(["imonth", "iday"], axis=1, inplace=True)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import TruncatedSVD
```

```python
numerical_features = [
    'extended', 'specificity', 'crit1', 'crit2', 'crit3', 'doubtterr', 'multiple', 
    'success', 'guncertain1', 'individual', 'nperps', 'nperpcap', 'claimed', 
    'claim2', 'claim3', 'nkill', 'nkillus', 'nkillter', 'nwound', 'nwoundus', 
    'nwoundte', 'property', 'propvalue', 'ishostkid', 'nhostkid', 'nhostkidus', 
    'nhours', 'ndays', 'ransom', 'ransomamt', 'ransomamtus', 'ransompaid', 
    'ransompaidus', 'nreleased', 'INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY'
]

categorical_features = [
    'resolution', 'country_txt', 'region_txt',
    'alternative_txt', 'attacktype1_txt',
    'targtype1_txt', 'targsubtype1_txt', 'corp1', 'target1',
    'natlty1_txt', 'targtype2_txt', 'targsubtype2_txt', 'corp2', 'target2',
    'natlty2_txt', 'targtype3_txt', 'targsubtype3_txt', 'corp3', 'target3',
    'natlty3_txt', 'gname', 'gname2', 'gname3', 'motive', 'claimmode_txt', 
    'claimmode2_txt', 'claimmode3_txt', 'weaptype1_txt', 'weaptype2_txt', 
    'weaptype3_txt', 'weaptype4_txt', 'propextent_txt', 'divert', 
    'kidhijcountry', 'hostkidoutcome_txt'
]
```

```python
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='Empty')), ('encoder', OneHotEncoder())
])
```

```python
preprocessing = ColumnTransformer(
    transformers=[
        ('num_trans', numerical_transformer, numerical_features),
        ('cat_trans', categorical_transformer, categorical_features)
    ]
)
```

```python
preprocessing.fit_transform(new_df)
```

```python
feature_selection = Pipeline(
    steps=[
        ('preprocessing', preprocessing),
        ('pca', TruncatedSVD(n_components=10000))
    ])
```

```python
feature_selection.fit_transform(new_df)
```
