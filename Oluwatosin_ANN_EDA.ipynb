```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region','attacktype1_txt':'AttackType1','attacktype2_txt':'AttackType2','attacktype3_txt':'AttackType3','targtype1_txt':'TargetType1','targsubtype1_txt':'TargetSubtype1','targtype2_txt':'TargetType2','targsubtype2_txt':'TargetSubtype2','targtype3_txt':'TargetType3', 'targsubtype3_txt':'TargetSubtype3', 'provstate':'State', 'target1':'Target1','nkill':'Killed','nwound':'Wounded','weaptype1':'WeaponType1', 'weaptype2':'WeaponType2','weaptype3':'WeaponType3','weaptype4':'WeaponType4','gname':'Group', 'summary':'Summary','city':'City','location':'Location', 'target2':'Target2','target3':'Target3'}, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
global_df=df[['Year','Month','Day','Country','Region','AttackType1','TargetType1','State','Target1','Killed','Wounded','WeaponType1','Group','City','Location','latitude','longitude','vicinity','claimed']]
```

```python
global_df.tail()
```

```python
global_df.isnull().sum()
```

```python
global_df['Wounded']=global_df['Wounded'].fillna(0)
```

```python
global_df['latitude']=global_df['latitude'].fillna(0)
global_df['Killed']=global_df['Killed'].fillna(0)
global_df['longitude']=global_df['longitude'].fillna(0)
global_df['claimed']=global_df['claimed'].fillna(0)
```

```python
global_df['State'].fillna('Unknown',inplace=True)
global_df['City'].fillna('Unknown',inplace=True)
global_df['Target1'].fillna('Unknown',inplace=True)
global_df['Location'].fillna('Unknown',inplace=True)
```

```python
global_df.isnull().sum()
```

```python
print('Country with the maximum number of attacks: ', global_df['Country'].value_counts().index[0])
print('Region with the maximum number of attacks: ', global_df['Region'].value_counts().index[0])
print('State with the maximum number of attacks: ', global_df['State'].value_counts().index[0])
print('Location with the maximum number of attacks: ', global_df['Location'].value_counts().index[0])
print('City with the maximum number of attacks: ', global_df['City'].value_counts().index[1])
print('Year with the maximum number of attacks: ', global_df['Year'].value_counts().index[0])
print('Month with the maximum number of attacks: ', global_df['Month'].value_counts().index[0])
print('Group with the maximum number of attacks: ', global_df['Group'].value_counts().index[1])
print('Most attack: ', global_df['AttackType1'].value_counts().index[0])


```

```python
plt.subplots(figsize=(25,6))
sns.countplot('Year',data=global_df)
plt.title('Number of terrorist activities each year')
plt.show()
```
