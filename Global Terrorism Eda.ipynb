```python
a=df1[['provstate','city','corp1','weapsubtype1_txt','weapdetail','targsubtype1_txt']]
for i in a.columns:
 df1[i].fillna('Unknown',inplace=True)
```

```python
df1['latitude'].fillna(value=df1['latitude'].median(), inplace=True)
df1['longitude'].fillna(value=df1['longitude'].median(), inplace=True)
df1['specificity'].fillna(value=df1['specificity'].median(), inplace=True)
df1['guncertain1'].fillna(value=df1['guncertain1'].median(), inplace=True)
df1['weapsubtype1'].fillna(value=df1['weapsubtype1'].median(), inplace=True)
df1['nkill'].fillna(value=df1['nkill'].median(), inplace=True)
df1['ishostkid'].fillna(value=df1['ishostkid'].median(), inplace=True)
```
